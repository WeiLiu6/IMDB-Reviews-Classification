```python
num_reviews = train['review'].size
clean_reviews = []
for i in range(num_reviews):
    # If the index is evenly divisible by 1000, print a message
    if (i+1)%1000 == 0:
        print("Review {0} of {1}\n".format(i, num_reviews))
    clean_reviews.append(to_words(train['review'][i]))
    
```

```python
print('Creating the bag of words ...\n')
from sklearn.feature_extraction.text import CountVectorizer
# Initialize the "CountVectorizer" object, which is scikit-learn's
# bag of words tool.  
vectorizer = CountVectorizer(analyzer='word', max_features=5000)
train_data_features = vectorizer.fit_transform(clean_reviews)
train_data_features = train_data_features.toarray()
```

```python
print(train_data_features.shape)
vocab = vectorizer.get_feature_names()
print(vocab)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=100)
forest.fit(train_data_features, train['sentiment'])
```

```python
test = pd.read_csv(path+"testData.tsv", header=0, delimiter='\t', quoting=3)
test.head()
```

```python
num_test = test.shape[0]
clean_test_reviews =[]
for i in range(num_test):
    # If the index is evenly divisible by 1000, print a message
    if (i+1)%1000 == 0:
        print("Review {0} of {1}\n".format(i, num_test))
    clean_test_reviews.append(to_words(test['review'][i]))
    
```

```python
test_data_features = vectorizer.transform(clean_test_reviews)
test_data_features = test_data_features.toarray()
result = forest.predict(test_data_features)
out_put = pd.DataFrame(data={'id':test['id'], 'sentiment':result})
out_put
```
